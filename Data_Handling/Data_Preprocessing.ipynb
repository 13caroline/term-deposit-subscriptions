# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import KNNImputer as knn
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing as sk_pp
from sklearn.model_selection import train_test_split

%matplotlib inline
```

## Get Data

```python
#data = pd.read_csv('../../Data_AA2/bank-additional-full.csv', sep = ';',na_values="unknown") marcar nulos
data = pd.read_csv('../../Data_AA2/bank-additional-full.csv', sep = ';')
data.head()
```

```python
data.info()
```

```python
def age(df):
    df.loc[df['age'] <= 19, 'age'] = 1
    df.loc[(df['age'] > 19) & (df['age'] <= 30), 'age'] = 2
    df.loc[(df['age'] > 30) & (df['age'] <= 60), 'age'] = 3
    df.loc[(df['age'] > 60) & (df['age'] <= 98), 'age'] = 4
           
    return df

age(data)
data.head()
```

```python
def duration(df):

    df.loc[df['duration'] <= 102, 'duration'] = 1
    df.loc[(df['duration'] > 102) & (df['duration'] <= 180)  , 'duration'] = 2
    df.loc[(df['duration'] > 180) & (df['duration'] <= 319)  , 'duration'] = 3
    df.loc[(df['duration'] > 319) & (df['duration'] <= 644.5), 'duration'] = 4
    df.loc[df['duration']  > 644.5, 'duration'] = 5

    return df

duration(data)
data.head()
```

# Model 1 - Every coloumn and normalization

```python
data_m1 = pd.read_csv('../../Data_AA2/bank-additional-full.csv', sep = ';')
pd.set_option('display.max_columns', 21)
data_m1.head()
```

```python
# Label encoder order is alphabetical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_m1['job'] = le.fit_transform(data_m1['job']) 
data_m1['marital'] = le.fit_transform(data_m1['marital']) 
data_m1['education'] = le.fit_transform(data_m1['education']) 
data_m1['default'] = le.fit_transform(data_m1['default']) 
data_m1['housing'] = le.fit_transform(data_m1['housing']) 
data_m1['loan'] = le.fit_transform(data_m1['loan'])
data_m1['contact'] = le.fit_transform(data_m1['contact']) 
data_m1['month'] = le.fit_transform(data_m1['month']) 
data_m1['day_of_week'] = le.fit_transform(data_m1['day_of_week'])
data_m1['poutcome'] = le.fit_transform(data_m1['poutcome'])

data_m1['y'].replace(['no', 'yes'], [0,1], inplace  = True)
data_m1.head(10)
```

```python
from sklearn.model_selection import train_test_split

data_m1_Y = data_m1['y']
data_m1 = data_m1.drop(columns=['y']) 

X_train_m1, X_test_m1, Y_train_m1, Y_test_m1 = train_test_split( data_m1, data_m1_Y, test_size=0.30)
```

```python
X_train_m1.head()
```

```python
Y_train_m1.value_counts()
```

```python
Y_test_m1.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_m1 = pd.DataFrame(scaler.fit_transform(X_train_m1), columns = X_train_m1.columns)
X_test_m1 = pd.DataFrame(scaler.transform(X_test_m1), columns = X_test_m1.columns)

X_train_m1.head(10)
```

```python
data_m5 = pd.concat([train_m5, test_m5], ignore_index=True)
```

```python
data_m5.info()
```

```python
data_m5.isnull().sum()
```

```python
null_job = data_m5['job'].isnull()
null_marital = data_m5['marital'].isnull()
null_education = data_m5['education'].isnull()
null_default = data_m5['default'].isnull()
null_housing = data_m5['housing'].isnull()
null_loan = data_m5['loan'].isnull()
```

```python
data_m5 = data_m5.fillna("?")
```

```python
# Label encoder order is alphabetical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_m5['job'] = le.fit_transform(data_m5['job']) 
data_m5['marital'] = le.fit_transform(data_m5['marital']) 
data_m5['education'] = le.fit_transform(data_m5['education']) 
data_m5['default'] = le.fit_transform(data_m5['default']) 
data_m5['housing'] = le.fit_transform(data_m5['housing']) 
data_m5['loan'] = le.fit_transform(data_m5['loan'])
data_m5['contact'] = le.fit_transform(data_m5['contact']) 
data_m5['month'] = le.fit_transform(data_m5['month']) 
data_m5['day_of_week'] = le.fit_transform(data_m5['day_of_week'])
data_m5['poutcome'] = le.fit_transform(data_m5['poutcome'])

data_m5['y'].replace(['no', 'yes'], [0,1], inplace  = True)
data_m5.head(10)
```

```python
data_m5.loc[null_job==True,'job'] = np.nan
data_m5.loc[null_marital==True,'marital'] = np.nan
data_m5.loc[null_education==True,'education'] = np.nan
data_m5.loc[null_default==True,'default'] = np.nan
data_m5.loc[null_housing==True,'housing'] = np.nan
data_m5.loc[null_loan==True,'loan'] = np.nan
```

```python
data_m5.isnull().sum()
```

```python
imputer = knn(n_neighbors=5,weights='uniform', metric='nan_euclidean')
```

```python
data_m5 = pd.DataFrame(np.round(imputer.fit_transform(data_m5)), columns = data_m5.columns)
```

```python
train_m5 = data_m5.iloc[:28831, :]
test_m5 = data_m5.iloc[28831:, :]
```

```python
train_m5_Y = train_m5['y']
train_m5 = train_m5.drop(columns=['y']) 

test_m5_Y = test_m5['y']
test_m5 = test_m5.drop(columns=['y']) 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

train_m5 = pd.DataFrame(scaler.fit_transform(train_m5), columns = train_m5.columns)
test_m5 = pd.DataFrame(scaler.transform(test_m5), columns = test_m5.columns)

train_m5.head(10)
```

```python
train_m5['y'] = train_m5_Y.values
test_m5['y'] = test_m5_Y.values

train_m5.head(10)
```

```python
test_m5 = test_m5.reset_index()
test_m5 = test_m5.drop(columns=['index'])
test_m5.head()
```

```python
test_m5.loc[null_job_test==True,'job'] = np.nan
test_m5.loc[null_marital_test==True,'marital'] = np.nan
test_m5.loc[null_education_test==True,'education'] = np.nan
test_m5.loc[null_default_test==True,'default'] = np.nan
test_m5.loc[null_housing_test==True,'housing'] = np.nan
test_m5.loc[null_loan_test==True,'loan'] = np.nan
```
