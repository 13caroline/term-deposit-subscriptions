# Logistic Regression

### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectPercentile, f_classif
```

### Model 1
#### Read CSV

```python
train_m1 = pd.read_csv('../../../../Data_AA2/train_m1.csv', sep = ',')
train_m1.head()
```

```python
test_m1 = pd.read_csv('../../../../Data_AA2/test_m1.csv', sep = ',')
test_m1.head()
```

```python
#get target
train_m1_target = train_m1['y']
train_m1 = train_m1.drop(columns=['y'])
```

```python
#get target
test_m1_target = test_m1['y']
test_m1 = test_m1.drop(columns=['y'])
```

#### Hyperparameter tuning using GridSearchCV

```python
# Create the parameter grid based on the results of random search 
parameters = {'C': [0.001,0.01,0.1,1,10,100,1000]}

# Create the model
logistic_regression_model = LogisticRegression()

# Best model
opt_model_lr = GridSearchCV(logistic_regression_model, parameters,  scoring='accuracy', verbose=1)

# Fit the model
opt_model_lr.fit(train_m1, train_m1_target)

print (opt_model_lr.best_estimator_)
```

#### Create and fit the model

```python
# Create model with best parameters
logistic_regression_model = LogisticRegression(C=0.01)

# Fit the best model
logistic_regression_model.fit(train_m1, train_m1_target)
logistic_regression_model
```

```python
predictions = logistic_regression_model.predict(test_m1)

# Confusion matrix
confusion = pd.crosstab(test_m1_target,predictions, rownames=['Actual'], colnames=['Predicted'], margins=True)
print(confusion )
```

```python
# Accuracy
score = accuracy_score(test_m1_target,predictions)
print('Accuracy:{0:f}'.format(score))
```

```python
# Report
print(classification_report(test_m1_target,predictions))
```

```python
# Validate model using best model and cross validation
pecc_lr = cross_val_score(logistic_regression_model, train_m1, train_m1_target, cv = 5).mean()
pecc_lr
```
